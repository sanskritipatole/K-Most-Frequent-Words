```python
# plotting graphs
# 300MB_Data
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
labels = ["Using Map-Reduce", "Using Map-Reduce with Combiner", "Using Map-Reduce with Partitioner", "Using Map-Reduce with Compression"]
x_axis = ["Technique 1", "Technique 2", "Technique 3", "Technique 4"]
y_axis = [62, 48, 51, 53]
plt.figure(figsize=(10,6))
plt.bar(x_axis, y_axis, color = ["red", "blue", "green", "yellow"])
plt.title('Experiment-1 300MB Dataset(top 100 words)')
plt.xlabel('Techniques')
plt.ylabel('Execution time in seconds')
EF = mpatches.Patch(color='red', label="Using Map-Reduce")
LL = mpatches.Patch(color='blue', label="Using Map-Reduce with Combiner")
PP1 = mpatches.Patch(color='green', label="Using Map-Reduce with Partitioner")
PP2 = mpatches.Patch(color='yellow', label="Using Map-Reduce with Compression")
plt.legend(handles=[EF, LL, PP1, PP2], loc="upper right")
plt.show()
```
